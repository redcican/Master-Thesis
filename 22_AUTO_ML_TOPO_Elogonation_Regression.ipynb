```python
df = pd.read_csv('Elongation_tiny_regression.csv')
df.head()
```

```python
X = np.array(df.values[:, :-1])
y = np.array((df['Elong']).values)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
tpot = TPOTRegressor(verbosity=2, max_time_mins=150, cv=10, n_jobs=-1)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
```

```python
feature_loc = ['C','SI','MN','P','S','CU','MG','CR','NI','PB','AL','MO','TI','SN','V','ZN','CE']
X = df.loc[:,feature_loc]
Y = df.Elong

print(X.shape, Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=.2, random_state=0)
```

```python
model = XGBRegressor(learning_rate=0.1,
                    max_depth=5,
                    subsampple=0.9,
                    max_features=0.45,
                     n_estimators=100,
                    min_samples_leaf=16,
                    min_samples_split=5,subsample=0.9)
```

```python
model.fit(X_train, y_train, eval_metric='mae',verbose=True,
         eval_set=[(X_test,y_test)],early_stopping_rounds=20)
```

```python
from sklearn.metrics import mean_absolute_error
mse = mean_absolute_error(y_test, model.predict(X_test))
print("MSE: %.4f" % mse)
```

```python
xgb.plot_importance(model, importance_type="weight")
plt.title('XGBoost plot importance (model, importance_type="weight")')
plt.show()
```

```python
xgb.plot_importance(model,importance_type="cover",show_values=False)
plt.title('XGBoost.plot importance (model, importance_type="cover)')
plt.show()
```

```python
xgb.plot_importance(model,importance_type="gain",show_values=False)
plt.title('XGBoost.plot importance (model, importance_type="gain)')
plt.show()
```
